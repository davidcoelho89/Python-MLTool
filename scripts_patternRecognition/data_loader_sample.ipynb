```python
# Import Libraries

import numpy as np           # Work with matrices (arrays)
import pandas as pd          # Work with Dataframes 

import data_loader
```

```python
# ECG and AUDIO

X,Y = data_loader.load_ecg_audio()

display(X.shape)
display(Y.shape)

# Visualize Data
XY = np.concatenate([X,Y]).T
df = pd.DataFrame(XY)
df.columns = ['minimum','maximum','mean','var','std','skew','kurt','iqr','label']
df.head()
```
